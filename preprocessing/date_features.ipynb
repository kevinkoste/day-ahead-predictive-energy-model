```python
import pandas as pd
import numpy as np
from sklearn import preprocessing

# import utility_functions as fn

import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
start = '2012-01-01 00:00:00'
end = '2019-03-01 00:00:00'

dates = pd.DataFrame(index=pd.DatetimeIndex(start=start, end=end, freq='H'))
```

```python
# adds hour of day features
for i in range(0,24):
    dates['hr_'+str(i)] = np.array([i]*len(dates))
    dates.loc[:,'hr_'+str(i)] = (dates.loc[:,'hr_'+str(i)].values == np.asarray(dates.index.hour)).astype(int)
```

```python
# adds day of week features
for i in range(0,7):
    dates['day_'+str(i)] = np.array([i]*len(dates))
    dates.loc[:,'day_'+str(i)]=(dates.loc[:,'day_'+str(i)].values==np.asarray(dates.index.dayofweek)).astype(int)
    
dates.head()
```

```python
dates.to_csv('../data/date_features.csv')
```
